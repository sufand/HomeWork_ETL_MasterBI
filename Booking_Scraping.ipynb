```python
def detail(element):
  url = element.find_elements_by_css_selector(".js-sr-hotel-link")[0].get_attribute("href")
  name = ""
  rating = ""
  geoloc = ""
  score = ""
  reviews = ""
  breakfast = ""
  canc_policy = ""
  payment_policy = ""
  price = ""
  try:
    if(len(element.find_elements_by_css_selector(".sr-hotel__name"))>0):
      name = element.find_elements_by_css_selector(".sr-hotel__name")[0].text
    if(len(element.find_elements_by_css_selector(".bui-rating"))>0):
      rating = element.find_elements_by_css_selector(".bui-rating")[0].get_attribute("aria-label")
    if(len(element.find_elements_by_css_selector(".bui-link"))>0):
      geoloc = element.find_elements_by_css_selector(".bui-link")[0].get_attribute("data-coords")
    if(len(element.find_elements_by_css_selector(".sr-review-score"))>0):
      score = element.find_elements_by_css_selector(".sr-review-score")[0].get_attribute("data-ugc-review-score")
    if(len(element.find_elements_by_css_selector(".sr-review-score"))>0):
      reviews = element.find_elements_by_css_selector(".sr-review-score")[0].get_attribute("data-ugc-review-nr")
    if(len(element.find_elements_by_css_selector(".sr_room_reinforcement"))>0):
      breakfast = element.find_elements_by_css_selector(".sr_room_reinforcement")[0].text
    if(len(element.find_elements_by_css_selector(".sr_room_reinforcement.e2e-free-cancellation"))>0):
      canc_policy = element.find_elements_by_css_selector(".sr_room_reinforcement.e2e-free-cancellation")[0].text
    if(len(element.find_elements_by_css_selector(".sr_room_reinforcement.e2e-prepayment"))>0):
      payment_policy = element.find_elements_by_css_selector(".sr_room_reinforcement.e2e-prepayment")[0].text
    if(len(element.find_elements_by_css_selector(".bui-price-display__value"))>0):
      price = element.find_elements_by_css_selector(".bui-price-display__value")[0].text
    hotel_id = url
  except:
    pass
  return {'url': url,
          'name': name,
          'rating': rating,
          'geoloc': geoloc,
          'score': score,
          'reviews': reviews,
          'breakfast': breakfast,
          'canc_policy': canc_policy,
          'payment_policy': payment_policy,
          'price': price,
          'hotel_id': hotel_id}
```

# Hotels - Dettaglio Pagina

```python
import pandas as pd
ds_hotels = pd.read_csv("hotels.csv", index_col="hotel_id")
ds_hotels.head()
```

```python
details = []
for hotel_id, hotel in tqdm(ds_hotels.iterrows(), total=len(ds_hotels)):
  link = hotel["url"]
  wd.set_window_size(1920, 1080)
  wd.get(link)
  time.sleep(3)
  facilities = []
  try:
    top_facilities = wd.find_element_by_css_selector(".hp_desc_important_facilities")
    for facility in top_facilities.find_elements_by_css_selector(".important_facility"):
      facilities.append(facility.text)
  except Exception as e:
    print(e)
  details.append({'url': link,
                  'facilities': facilities})
len(details)
```
